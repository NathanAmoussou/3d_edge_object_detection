```python
from ultralytics import YOLO

# 1. Charger le mod√®le
# Le fichier "yolo11n.pt" sera t√©l√©charg√© automatiquement s'il n'est pas pr√©sent
model = YOLO("yolo11n.pt")

# 2. Lancer l'inf√©rence sur une source (image, vid√©o, webcam ou URL)
# Remplacer 'bus.jpg' par le chemin de votre image ou 0 pour la webcam
results = model("https://ultralytics.com/images/bus.jpg")

# 3. Afficher ou sauvegarder les r√©sultats
for result in results:
    result.show()  # Affiche l'image √† l'√©cran
    result.save(filename="resultat.jpg")  # Sauvegarde l'image annot√©e
```

```python
from ultralytics import YOLO

# 1. Charger le mod√®le
model = YOLO("yolo11n.pt")

# 2. Exporter en format OpenVINO
# imgsz=416 ou 640. 
# 416x416 est recommand√© pour une bonne vitesse (FPS) sur OAK-D. 
# 640x640 sera plus pr√©cis mais plus lent (~15-20 FPS).
model.export(format="openvino", dynamic=False, half=True, imgsz=416)
```

```python
import blobconverter

blob_path = blobconverter.from_openvino(
    xml="yolo11n_openvino_model/yolo11n.xml",
    bin="yolo11n_openvino_model/yolo11n.bin",
    data_type="FP16",
    shaves=6,  # Utilise 6 coeurs du VPU
    use_cache=False,
    version="2022.1" # Version OpenVINO compatible
)

print(f"Mod√®le converti sauvegard√© ici : {blob_path}")
# Renommez le fichier obtenu en "yolo11n.blob" pour simplifier
```
